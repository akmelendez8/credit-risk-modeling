# Machine Learning Credit Risk Model for Predicting Loan Default
### Executive Overview

---

## Business Objective

This project develops and validates a machine learning model that predicts the probability a borrower will default on a loan using the LendingClub peer-to-peer lending dataset (2007 to 2018). Accurate default prediction supports underwriting decisions, risk-based pricing, and expected credit loss estimation under CECL (Probability of Default x Loss Given Default). Traditional tools like FICO scores evaluate risk factors independently and miss complex, nonlinear interactions between borrower characteristics that machine learning can capture.

## Data and Approach

200,000 loans were filtered to final outcomes: Fully Paid (approximately 80%) and Charged Off (approximately 20%). Nine raw variables were transformed into a 13-feature set through log transformations, a loan-to-income leverage ratio, and categorical encoding. Features were standardized using a training-set-only scaler to prevent data leakage, and the data was split 80/20 using stratified sampling.

## Models Evaluated

Four models were developed. **Logistic Regression** served as the interpretable baseline, confirming interest rate and FICO score as the strongest individual predictors; it meets regulatory interpretability requirements for adverse action notices. **Random Forest** averaged 200 decorrelated decision trees to capture nonlinear interactions and reduce variance. **XGBoost** built trees sequentially to correct prior errors, with built-in regularization, and achieved the highest AUC among individual models. **K-Means Clustering** segmented borrowers into four risk profiles without using the default label, revealing meaningfully different default rates useful for portfolio management and pricing.

## Tuning and Ensemble

All three supervised models were tuned using RandomizedSearchCV with 5-fold stratified cross-validation, optimizing for AUC-ROC. Classification thresholds were set below 0.5 using precision-recall curves to reflect the higher business cost of a missed default relative to a false positive. A soft-voting ensemble of the three tuned models provided incremental lift where component errors were sufficiently diverse.

## Results and Recommendation

Tuning produced consistent AUC improvement over baseline configurations for all three models. The **tuned XGBoost model** is recommended for primary deployment: it offers the highest discriminative power, native handling of class imbalance, and interpretable feature importance. Logistic Regression remains preferred in any context requiring regulatory explainability. Priority next steps include transitioning to a survival model for CECL lifetime loss estimation and incorporating macroeconomic variables for scenario analysis. Model calibration via Platt Scaling is recommended to ensure predicted probabilities reflect empirical default rates, which is required for direct use in CECL loss reserving. The held-out test evaluation and model documentation provide the validation evidence needed to support deployment in a governed lending environment.
